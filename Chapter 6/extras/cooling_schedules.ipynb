```python
import math
import numpy as np
import matplotlib.pyplot as plt

T_init=100 # initial temperature
T_fin=0.001 # final temperature

# Linear cooling schedule
N=10 # maximum number of iterations
T_lin=[]
beta=(T_init-T_fin)/N

#  Geometric cooling schedule
T_geo1=[]
T_geo2=[]
alpha1=0.7 # alpha values between 0.7 to 0.95 are commonly used
alpha2=0.9 

#  Logarithmic cooling schedule
T_log=[]
alpha=0.9 

#  Exponential cooling schedule
T_exp=[]



for i in range(1,N+1):
    T_lin.append(T_init-beta*i)
    T_geo1.append(T_init*alpha1**i)
    T_geo2.append(T_init*alpha2**i)
    T_log.append(T_init*alpha/math.log(1+i))
    T_exp.append(T_init*math.exp((-alpha*i)))
    
# Plot the annealing schedules
plt.figure(figsize=(10, 10))
plt.xticks(np.arange(0, N+1, step=1))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(list(range(1,N+1)), T_lin, linewidth=2, color='gray')
plt.plot(list(range(1,N+1)), T_geo1, linewidth=2, color='orange')
plt.plot(list(range(1,N+1)), T_geo2, linewidth=2, color='yellow')
plt.plot(list(range(1,N+1)), T_log, linewidth=2, color='blue')
plt.plot(list(range(1,N+1)), T_exp, linewidth=2, color='red')
plt.xlabel('iterations', fontsize=15, loc="right")
plt.ylabel('Temperature', fontsize=15)
plt.legend(['Linear Cooling schedule', r'Geometric Cooling: $\alpha$ =0.7', r'Geometric Cooling: $\alpha$=0.9', r'Logarithmic Cooling: $\alpha$=0.9', r'Exponential Cooling: $\alpha$=0.9'], fontsize=18);
plt.title('SA Cooling Schedules', fontsize=22)
plt.grid()
plt.savefig('annealing_schedules.png')
```
