```python
from scipy.optimize import dual_annealing
import matplotlib.pyplot as plt
import numpy as np

# define function 
def func(x):
    return (x-6)**2 

# define ranges
bounds = [(0, 31)] # boundary constraints
x = np.arange(bounds[0][0], bounds[0][1], 0.01)
f = func(x)

# perform the dual annealing search
res = dual_annealing(func, bounds=bounds, maxiter = 500) 
print(res)

plt.figure(figsize=(12, 10))
plt.title("Continuous function optimization using dual annealing")
plt.plot(x, func(x), label="f(x)")
plt.plot(res['x'], res['fun'], 'sr', label="global minimum")
print("global minimum: x = %.4f, f(x) = %.4f" % (res['x'], res['fun']))
plt.xlabel("x", fontsize=18)
plt.ylabel("f(x)", fontsize=18)
plt.legend(loc='best', fancybox=True, shadow=True, fontsize=18)
plt.grid()
plt.savefig('func_opt.png')
```
