## Generate and plot Gaussian distributions with different variance

```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# define standard deviations
variances = [0.3, 0.5, 0.8]

# x-axis for the plot
x_axis = arange(-5, 5, 0.001)
pyplot.figure(figsize=(8, 6))
# plot low variance
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.4), label=r"$\sigma^2$ = {}".format(variances[0]),linestyle='-')
# plot medium variance
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.8), label=r"$\sigma^2$ = {}".format(variances[1]),linestyle='--')
# plot high variance
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 1.5), label=r"$\sigma^2$ = {}".format(variances[2]),linestyle=':')
pyplot.grid(color='gray', linestyle='-', linewidth=0.4)
pyplot.xticks(arange(-5, 6, 1))
pyplot.legend(fontsize="10")
pyplot.xlabel('x')
# pyplot.show()
pyplot.savefig('CH08_Gaussians.png', format='png', dpi=300)
```
