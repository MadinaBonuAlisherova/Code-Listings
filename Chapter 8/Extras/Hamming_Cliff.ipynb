### Calculating and visualizing Hamming distance to understand Hamming Cliff

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to convert decimal to binary
def decimal_to_binary(n, num_bits):
    return format(n, f'0{num_bits}b')

# Function to convert decimal to Gray code
def decimal_to_gray(n, num_bits):
    n = n ^ (n >> 1)
    return format(n, f'0{num_bits}b')

# Function to calculate Hamming distance
def hamming_distance(s1, s2):
    return sum(c1 != c2 for c1, c2 in zip(s1, s2))

decimal_values = np.arange(17)
num_bits = 4

binary_codes = [decimal_to_binary(n, num_bits) for n in decimal_values]
gray_codes = [decimal_to_gray(n, num_bits) for n in decimal_values]

binary_hamming_distances = [hamming_distance(binary_codes[i], binary_codes[i+1]) for i in range(len(binary_codes)-1)]
gray_hamming_distances = [hamming_distance(gray_codes[i], gray_codes[i+1]) for i in range(len(gray_codes)-1)]

plt.plot(decimal_values[:-1], binary_hamming_distances, 'o-', label="Binary Coding")
plt.plot(decimal_values[:-1], gray_hamming_distances, 's-', label="Gray Coding")
plt.xticks(decimal_values[:-1])
plt.xlabel("Decimal Value")
plt.ylabel("Hamming Distance")
plt.legend()
plt.grid()
# plt.show()
plt.savefig('CH08_F02_Hamming Cliff.png', format='png', dpi=300)
```
