# Different inertia weight update methods

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
w_max = 0.9 # maximum inertia weight
w_min = 0.4 # minimum inertia weight
t_max = 100 # number of iterations

# Time steps
t = np.arange(0, t_max)

# Random Selection (RS)
w_rs = 0.5 + np.random.rand(t_max) / 2

# Linear Time Varying (LTV)
w_ltv = (w_max - w_min) * (t_max - t) / t_max + w_min

# Non-Linear Time Varying (NLTV)
w_nltv = np.zeros(t_max)
w_nltv[0] = w_max
for i in range(1, t_max):
    w_nltv[i] = (w_nltv[i-1] - 0.4) * ((t_max - i) / (t_max + 0.4))

# Plot
plt.figure(figsize=(10, 8))
plt.plot(t, w_rs, label='Random Selection (RS)', linestyle='--', color='g')
plt.plot(t, w_ltv, label='Linear Time Varying (LTV)', linestyle='-', color='r')
plt.plot(t, w_nltv, label='Non-Linear Time Varying (NLTV)', linestyle=':', color='b')

plt.xlabel('Time', fontsize=14)
plt.ylabel('Inertia weight', fontsize=14)
plt.legend(loc='lower right', fontsize=14)
plt.show()
# plt.savefig('CH09_F13_inertia.png', dpi=300)
```

# Cognitive and social acceleration coefficient updates

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the parameters
c1_max = 2.0 # maximum cognitive acceleration coefficient
c1_min = 0.5 # minimum cognitive acceleration coefficient
c2_max = 2.5 # maximum social acceleration coefficient
c2_min = 0.8 # minimum social acceleration coefficient
t_max = 100 # number of iterations

# Define the functions
def c1(t):
    return (((c1_max - c1_min) * (t_max - t)) / t_max) + c1_min

def c2(t):
    return c2_max - (((c2_max - c2_min) * (t_max - t)) / t_max)

# Generate the x-values
t_values = np.arange(0, t_max+1)

# Generate the y-values for the two functions
c1_values = [c1(t) for t in t_values]
c2_values = [c2(t) for t in t_values]

# Plot the two functions
plt.figure(figsize=(10, 8))
plt.plot(t_values, c1_values, label='c1(t)', linestyle='-', color='r')
plt.plot(t_values, c2_values, label='c2(t)', linestyle='--', color='b')

# Add axis labels and a legend
plt.xlabel('Time', fontsize=14)
plt.ylabel('Cognitive and Social Acceleration Coefficients', fontsize=14)
plt.legend(fontsize=14)

# Show the plot
plt.show()
# plt.savefig('CH09_F14_acceleration.png', dpi=300)
```
