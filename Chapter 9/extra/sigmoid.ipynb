```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# create a range of x values
x = np.linspace(-4, 4, 100)

# compute sigmoid function values for each x value
y = sigmoid(x)

# plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, y)

plt.xlabel('$v_k^{id}$', fontsize=14)
plt.ylabel('$sig(v_k^{id})$', fontsize=14)

# add vertical and horizontal lines at v=0.0
plt.axvline(x=0.0, linestyle='--', color='gray')
plt.axhline(y=sigmoid(0.0), linestyle='--', color='gray')

# show the value of the sigmoid function at v=0.0
y_05 = sigmoid(0.0)
plt.text(0.0, y_05, f'({0.0}, {y_05:.2f})', ha='left', va='bottom')

# add the LaTeX equation
eqn = r'prob($x_{k+1}^{id}=1$)=' + '\n0.5 if $v_k^{id}=0$,' + '\n <0.5 if $v_k^{id}<0$,' + '\n >0.5 if $v_k^{id}>0$'
plt.text(-4, 0.71, eqn, fontsize=12, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))

plt.show()
# plt.savefig('sigmoid.png', dpi=300, bbox_inches='tight')
```
