```python
import math
import numpy as np
import matplotlib.pyplot as plt

T_init=100 # initial temperature
T_fin=0.001 # final temperature

# Linear cooling schedule
N=100 # maximum number of iterations
T_lin=[T_init]
beta=(T_init-T_fin)/N

#  Geometric cooling schedule
T_geo1=[T_init]
T_geo2=[T_init]
alpha1=0.7 # alpha values between 0.7 to 0.95 are commonly used
alpha2=0.9

#  Logarithmic cooling schedule
alpha_log = 10
T_log=[T_init]

alpha=0.9
#  Exponential cooling schedule
T_exp=[T_init]
T_lin_inv=[T_init]



for i in range(1,N+1):
    T_lin.append(T_init-beta*i)
    T_geo1.append(T_init*alpha1**i)
    T_geo2.append(T_init*alpha2**i)
    T_log.append(T_init/(1+alpha_log*math.log(1+i)))
    T_exp.append(T_init*math.exp((-alpha*i)))
    T_lin_inv.append(T_init/(1+alpha*i))

# Plot the annealing schedules
plt.figure(figsize=(10, 10))
plt.xticks(np.arange(0, N+1, step=N/10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(list(range(0,N+1)), T_lin,     '-', linewidth=2, color='gray')
plt.plot(list(range(0,N+1)), T_geo1,    '-*', markersize=10, linewidth=2, color='orange')
plt.plot(list(range(0,N+1)), T_geo2,    '-v', markersize=10, linewidth=2, color='magenta')
plt.plot(list(range(0,N+1)), T_log,     '--', markersize=10, linewidth=2, color='blue')
plt.plot(list(range(0,N+1)), T_exp,     '-+', markersize=10, linewidth=2, color='red')
plt.plot(list(range(0,N+1)), T_lin_inv, '-.', markersize=10, linewidth=2, color='green')
plt.xlabel('iterations', fontsize=15, loc="right")
plt.ylabel('Temperature', fontsize=15)
plt.legend(['Linear Cooling schedule', r'Geometric Cooling: $\alpha$='+str(alpha1), r'Geometric Cooling: $\alpha$='+str(alpha2), r'Logarithmic Cooling: $\alpha$='+str(alpha_log), r'Exponential Cooling: $\alpha$='+str(alpha), r'Linear Inverse: $\alpha$='+str(alpha)], fontsize=18);
plt.title('SA Cooling Schedules', fontsize=22)
plt.grid()
plt.savefig('annealing_schedules.png')
```
