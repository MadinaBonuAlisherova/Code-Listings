# Gaussian Neighborhood Function in SOM

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, sigma):
    return np.exp(-x**2 / (2*sigma**2))

# Create a range of values
x = np.linspace(-3, 3, 100)

# Specify a value for sigma
sigma = 0.61

# Calculate the Gaussian values
y = gaussian(x, sigma)

# Plot the Gaussian function
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Gaussian Neighborhood Function')
plt.xlabel('Distance from the winning neuron $LDist(u,v)$')
plt.ylabel('Influence of the winning neuron $\Theta (u,v)$')
plt.grid(True)
plt.show()
```
