## Celestial Coordinate System

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and an axis
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='polar')

# Set the grid and labels for Right Ascension (RA)
ax.set_theta_direction(-1)
ax.set_theta_zero_location('N')
ax.set_xticks(np.radians(np.linspace(0, 360, 24, endpoint=False))) # 24 hours
ax.set_xticklabels([f'{int(hour)}h' for hour in np.linspace(0, 24, 24, endpoint=False)])

# Set the grid and labels for Declination (Dec)
ax.set_ylim(-90, 90)
ax.set_yticks(np.linspace(-90, 90, 13)) # every 15 degrees
ax.set_yticklabels([f'{int(deg)}°' for deg in np.linspace(-90, 90, 13)])

# Set labels
ax.set_xlabel("Right Ascension")
ylabel = ax.set_ylabel("Declination", va='bottom', labelpad=40.0)  # Adjust the labelpad for positioning

ax.grid(True)

# Add a title
ax.set_title("Celestial Coordinate System")

# Example Point (RA in hours, Dec in degrees)
# Convert RA from hours to degrees (1 hour = 15 degrees)
ra_example = 10 * 15  # 10 hours
dec_example = 30  # 30 degrees

# Convert RA to radians for the plot
ra_example_rad = np.radians(ra_example)

# Plot the example point
ax.plot(ra_example_rad, dec_example, 'o', color='red')

# Label the point
label = f'(10h, 30°)'
ax.text(ra_example_rad, dec_example, label, horizontalalignment='left', verticalalignment='bottom')

# Show the plot
plt.show()

# Save the plot
plt.savefig('CH11_F15_Celestial_system.png', dpi=300, bbox_inches='tight')
```
