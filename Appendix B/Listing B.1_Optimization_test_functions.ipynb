## Many Local Minima Functions

### [Schwefel Function](https://www.sfu.ca/~ssurjano/schwef.html)

The Schwefel function is complex, with many local minima. The plot shows the two-dimensional form of the function.

Formula: $f(x)=418.9829d+\sum_{i=1}^{d}x_isin(\sqrt|x_i|)$ , $x_i\in [-500,500]$

Dimensions: d

Input Domain: The function is usually evaluated on the hypercube $x_i ∈ [-500, 500]$, for all $i = 1, …, d$.

Global Minimum: $f(x^*)=0$, at $x^* = (420.9687,…,420.9687)$

```python
#!pip install sympy
import sympy as sy
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=418.9829*2 - x1 * sy.sin(sy.sqrt( abs( x1 )))-x2*sy.sin(sy.sqrt(abs(x2)))

plot3d(f, (x1, -500, 500), (x2,-500, 500))
```

### [Ackley Function](https://www.sfu.ca/~ssurjano/ackley.html)

The Ackley function is a popular benchmark function used to test optimization algorithms. It is is defined in n-dimensional space and is typically used for problems involving continuous optimization.

Formula: $f(x)=-a \exp(-b \sqrt(1/d \sum_{i=1}^{d}x_i^2) - b \exp(1/d \sum_{i=1}^{d}cos(cx_i))  + a + \exp(1)$

Dimensions: d

Input Domain: The function is usually evaluated on the hypercube $x_i ∈ [-32.768, 32.768]$, for all $i = 1, …, d$.

Global Minimum: $f(x^*)=0$, at $x^* = (0,…,0)$

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot3d
import numpy as np

x1, x2 = symbols('x1 x2')

a = 20
b = 0.2
c = 2 * np.pi

f= -a * sy.exp(-b * sy.sqrt(0.5 * (x1**2 + x2**2))) -sy.exp(0.5 * (sy.cos(c * x1) + sy.cos(c * x2))) + a + sy.exp(1)

plot3d(f, (x1, -32.768, 32.768), (x2,-32.768, 32.768))
```

### [Bukin Function N. 6](https://www.sfu.ca/~ssurjano/bukin6.html)

Formula: $f(x_1,x_2)=100\sqrt|x_2-0.01{x_1}^2| + 0.01|x_1 + 10|$

Dimensions: 2

Input Domain: The function is usually evaluated on the rectangle $x_1 \in [-15, -5]$, $x_2 \in [-3, 3]$

Global Minimum: The sixth Bukin function has many local minima, all of which lie in a ridge

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=100*sy.sqrt(abs(x2-0.01*x1*x1)) +0.01*abs(x1+10)

plot3d(f, (x1, -15, -5), (x2,-3, 3))
```

### [Gramacy \& Lee function](https://www.sfu.ca/~ssurjano/grlee12.html)

This is a simple one-dimensional test function

Formula: $f(x)=\frac{sin(10\pi x)}{2x} + (x-1)^4$

Dimensions: 1

Input Domain: $x\in [0.5,2.5]$

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')

f=sy.sin(10*sy.pi*x)/(2*x)+pow((x-1),4)

p1 = plot(f, (x,0.5,2.5), show=False)
p1.show()
```

### [Griewank 1D function](https://www.sfu.ca/~ssurjano/griewank.html)

The Griewank function has many widespread local minima, which are regularly distributed.

Formula: $f(x)=\sum_{i=1}^{d} \frac{{x_i}^2}{4000} + \prod_{i=1}^{d} cos(\frac{x_i}{\sqrt{i}}) +1 $

Dimensions: d

Input Domain: $x\in [-600, 600]$

Global Minimum: $f(x^*)=0$, at $x^*=(0,...,0)$

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')

f=x**2/4000+sy.cos(x)+1

p1 = plot(f, (x, -200, 200))
```

## Bowl-Shaped Functions

### [Bohachevsky function](https://www.sfu.ca/~ssurjano/boha.html)

Formula: $f(x_1,x_2)={x_1}^2 +2{x_2}^2-0.3cos(3\pi x_1)-0.4cos(4\pi x_2)+0.7$

Dimensions: 2

Input Domain: $x_1\in [-100,100]$ and $x_2\in [-100,100]$

Global Minimum: $f(x^*)=0$, at $x^*=(0,0)$

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=x1**2 +2*(x2**2)-0.3*sy.cos(3*(sy.pi)*x1)-0.4*sy.cos(4*(sy.pi)*x2)+0.7
   
plot3d(f, (x1, -100, 100), (x2,-100, 100))
```

## Plate-Shaped Function 

### [McCormick Function](https://www.sfu.ca/~ssurjano/mccorm.html)

Formula: $f(x_1,x_2)=sin(x_1+x_2)+(x_1-x_2)^2-1.5x_1+2.5x_2+1$

Dimensions: 2

Input Domain: The function is usually evaluated on the rectangle $x_1\in [-1.5,4]$ and $x_2\in [-3,5]$ 

Global Minimum: $f(x^*)=-1.9133$, at $x^* = (-0.54719, -1.54719)$

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=sy.sin(x1+x2)-(x1-x2)**2-1.5*x1+2.5*x2+1

plot3d(f, (x1, -1.5, 4), (x2,-3, 5))
```

### [Booth Function](https://www.sfu.ca/~ssurjano/booth.html)

Formula: $f(x_1,x_2)=(x_1+2x_2-7)^2 +(2x_1+x_2-5)^2$

Dimensions: 2

Input Domain: $x_1,x_2\in [-10,10]$

Global Minimum: $f(x^*)=0$, at $x^*=(1,3)$

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=(x1+2*x2-7)**2+(2*x1+x2-5)**2

plot3d(f, (x1, -10, 10), (x2,-10, 10))
```

### [Matyas Function](https://www.sfu.ca/~ssurjano/matya.html)

Formula: $f(x)=0.26(x_1^2+x_2^2) -0.48x_1x_2$

Dimensions: 2

Input Domain: $x_1,x_2\in [-10,10]$

Global Minimum: $f(x^*)=0$, at $x^*=(0,0)$

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=0.26*(x1**2+x2**2)-0.48*x1*x2

plot3d(f, (x1, -10, 10), (x2,-10, 10))
```

## Valley-Shaped Functions

### [Three-Hump Camel Function](https://www.sfu.ca/~ssurjano/camel3.html)

Formula: $f(x)=2x_1^2-1.05x_2^4+\frac{{x_1}^6}{6}+x_1x_2+x_2^2$

Dimensions: 2

Input Domain: $x_1,x_2\in [-5,5]$

Global Minimum: $f(x^*)=0$, at $x^*=(0,0)$

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=2*x1**2-1.05*x1**4+(x1**6)/6+x1*x2+x2**2

plot3d(f, (x1, -5, 5), (x2, -5, 5))
```

### [Six-Hump Camel Function](https://www.sfu.ca/~ssurjano/camel6.html)

Formula: $f(x)=(4-2.1x_1^2+\frac{4x_1 ^4)}{3})x_1^2+x_1x_2+(-4+4x_2^2)x_2^2$

Dimensions: 2

Input Domain: $x_1\in [-3,3]$ and $x_2\in [-2,2]$

Global Minimum: $f(x^*)=-1.0316$, at $x^*=(0.0898,-0.7126)$ and $(-0.0898,0.7126)$

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=(4-2.1*x1**2+(4*x1**4)/3)*x1**2+x1*x2+(-4+4*x2**2)*x2**2

plot3d(f, (x1, -3, 3), (x2, -2, 2))
```

## Other

### [Beale Function](https://www.sfu.ca/~ssurjano/beale.html)

Formula: $f(x)=(1.5-x_1+x_1x_2)^2+(2.25-x_1+x_1x_2^2)^2+(2.625-x_1+x_1x_2^3)^2$

Dimensions: 2

Input Domain: $x_1,x_2\in [-4.5,4.5]$

Global Minimum: $f(x^*)=0$, at $x^*=(3,0.5)$

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=(1.5-x1+x1*x2)**2+(2.25-x1+x1*x2**2)**2+(2.625-x1+x1*x2**3)**2

plot3d(f, (x1, -4.5, 4.5), (x2, -4.5, 4.5))
```

## [Dixon-Price Function](https://www.sfu.ca/~ssurjano/dixonpr.html)

Formula: $f(x)=(x_1-1)^2+\sum_{i=2}^{d}i(2x_i^2-x_{i-1})^2$

Dimensions: d

Input Domain: $x_1,x_2,…,x_d\in [-10,10]$

Global Minimum: $f(x_i)=2^{- \frac{2^i-2}{2^i}}$, for $i=1,...,d$

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=(x1-1)**2+2*(2*x2**2-x1)**2

plot3d(f, (x1, -10, 10), (x2, -10, 10))
```

## [Styblinski-Tange Function](https://www.sfu.ca/~ssurjano/stybtang.html)

Formula: $f(x)=\frac{1}{2} \sum_{i=1}^{d}(x_i^4-16{x_i}^2+5x_i$

Dimensions: d

Input Domain: $x_1,x_2,…,x_d\in [-5,5]$

Global Minimum: $f(x^*)=-39.16599d$, at $x^*=(-2.903534,...,-2.903534)$

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=0.5 * ((x1**4 + x2**4) - 16 * (x1**2 + x2**2)+ 5 * (x1 + x2))

plot3d(f, (x1, -5, 5), (x2, -5, 5))
```

## [Forrwater et al.(2008) Function](https://www.sfu.ca/~ssurjano/forretal08.html)

Multimodal function with one global minimum, one local minimum and a zero-gradient inflection point.

Formula: $f(x)=(6x-2)^2sin(12x-4)$

Dimensions: 1

Input Domain: $x\in [0,1]$

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot

x = symbols('x')

f=(6*x-2)**2*sy.sin(12*x-4)

p1 = plot(f, (x, 0, 1))
```

## [Himmelblau's function](https://en.wikipedia.org/wiki/Himmelblau%27s_function) 

Formula: $f(x,y)=(x^2+y-11)^2+(x+y^2-7)^2$

Dimensions: 2

Input Domain: $x_1, x_2 \in [-6,6]$

Global Minimum: The Himmelblau function has four identical local minima at:
* f(3.0, 2.0)=0
* f(−2.805118, 3.131312)=0
* f(−3.779310, −3.283186)=0
* f(3.584428, −1.848126)=0

```python
from sympy import symbols
from sympy.plotting import plot3d

x1, x2 = symbols('x1 x2')

f=(x1**2+x2-11)**2 + (x1+x2**2-7)**2

plot3d(f, (x1, -6, 6), (x2,-6, 6))
```

## Mishra's Bird Function

Formula: $f(x,y)=sin(x)e^{(1-cos(y))^2}+cos(y)e^{(1-sin(x))^2}+(x-y)^2$. Subject to this constraint: $(x+5)^2+(y+5)^2<25$

Dimensions: 2

Input Domain: he search domain is usually: $-10 \le x \le 0$ and $-6.5 \le y \le 0$

Global Minimum: This function has two global minima at f(x*)= −106.764537 located at (4.70104 ,3.15294) and (−1.58214 ,−3.13024).

```python
import sympy as sy
from sympy import symbols
from sympy.plotting import plot3d


x1, x2 = symbols('x1 x2')

f=sy.sin(x1)*(sy.exp(1-sy.cos(x2))**2)+sy.cos(x2)*(sy.exp(1-sy.sin(x1))**2)+(x1-x2)**2

plot3d(f, (x1, -2*sy.pi,2*sy.pi,100), (x2,-2*sy.pi,2*sy.pi,100))
```

```python
# Example of how to import and visualize kursawe function
from deap import benchmarks
import numpy as np
import matplotlib.pyplot as plt

X=np.linspace(-5,10,100)
Y=np.linspace(-5,10,100)
X,Y=np.meshgrid(X,Y)
Z1=np.zeros(X.shape)
Z2=np.zeros(X.shape)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z1[i,j], Z2[i,j]=benchmarks.kursawe([X[i,j],Y[i,j]])

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(X,Y,Z1, cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Kursawe f1')

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,2,2, projection='3d')
ax.plot_surface(X,Y,Z2, cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Kursawe f2')

plt.show()
```

## pymoo Benchmarks

[pymoo: Multi-objective Optimization in Python](https://pymoo.org/problems/test_problems.html) also provides several optimization test functions.

```python
from pymoo.problems import get_problem
from pymoo.visualization.fitness_landscape import FitnessLandscape

# create a Ackly test problem
problem = get_problem("ackley", n_var=2, a=20, b=1/5, c=2 * np.pi)

# plot the fitness landscape
FitnessLandscape(problem, angle=(45, 45), _type="surface").show()
```

```python
# plot the fitness contour
FitnessLandscape(problem, _type="contour", colorbar=True).show()
```

# PySwarms Functions


[PySwarms](https://pyswarms.readthedocs.io/en/latest/api/pyswarms.utils.functions.html) implements the following test functions for optimization:

Function list: - Ackley’s, ackley - Beale, beale - Booth, booth - Bukin’s No 6, bukin6 - Cross-in-Tray, crossintray - Easom, easom - Eggholder, eggholder - Goldstein, goldstein - Himmelblau’s, himmelblau - Holder Table, holdertable - Levi, levi - Matyas, matyas - Rastrigin, rastrigin - Rosenbrock, rosenbrock - Schaffer No 2, schaffer2 - Sphere, sphere - Three Hump Camel, threehump

```python
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
import matplotlib.pyplot as plt
from pyswarms.utils.plotters import plot_cost_history

# Set-up hyperparameters
options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}

# Call instance of PSO
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options)

# Perform optimization
cost, pos = optimizer.optimize(fx.rastrigin, iters=1000)
```

```python
plot_cost_history(cost_history=optimizer.cost_history)
plt.show()
```
