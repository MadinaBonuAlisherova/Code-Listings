```python
import numpy
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt


# Objective function, converted into minimization
def objective(x):
    return 200 * x * x - 92000 * x + 8400000


# Convert back to positive values,
# divide by 1000 for visualization
def profit(x):
    return -1 * objective(x) / 1000


# Define the boundary constraints
x_min, x_max = 50, 340
x = numpy.arange(x_min, x_max, 0.1)

# Minimize the function
f = minimize_scalar(objective, method="brent")

opt_x, opt_f = f["x"], profit(f["x"])

print("Optimal Bike Price ($): %.2f" % opt_x)
print("Resultant Profit f(x) in K$: %.2f" % opt_f)
print("Total Evaluations n: %d" % f["nfev"])

# Plot inputs vs. target
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(x, profit(x), "-")

# Plot the optima
plt.plot([opt_x], [opt_f], "s", color="r")

# Add title and axis names
plt.xlabel("Bike price ($)", fontsize=20)
plt.ylabel("Profit (K$)", fontsize=20)
plt.show()
```
