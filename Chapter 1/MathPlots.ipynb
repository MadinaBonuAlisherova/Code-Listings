```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, num=200)

# A multimodal function
f = 3 * np.cos(x * x) - np.sin(3 * x) * np.exp(-x)

# Plot formatting
fig = plt.figure(figsize=(15, 15))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(x, f, linewidth=2)
plt.xlabel("x", fontsize=18, loc="right")
plt.ylabel("f(x)", fontsize=18)

# Draw a vertical line at the lower
# and upper bounds of the decision variable

plt.axvline(x=-3, ls="--", color="red")
plt.axvline(x=3, ls="--", color="red")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2)

# A convex function
f = x**2

# Use subplots for multiple graphs
fig, axs = plt.subplots(2, 2, sharex=True, figsize=(15, 15))

# The base function
axs[0, 0].plot(x, f, linewidth=2)
axs[0, 0].set_title("Base function $x^2$", fontsize=18)
axs[0, 0].set_ylim([-1, 5])

# Concave function
axs[1, 0].plot(x, -f, linewidth=2)
axs[1, 0].set_title("Duality $-x^2$", fontsize=18)
axs[1, 0].set_ylim([-5, 1])

# Stretched function
axs[0, 1].plot(x, f * 0.5, linewidth=2)
axs[0, 1].set_title("Stretched $\dfrac{x^2}{2}$", fontsize=18)
axs[0, 1].set_ylim([-1, 5])

# Shifted function
axs[1, 1].plot(x, f + 3, linewidth=2)
axs[1, 1].set_title("Shifted $x^2+3$", fontsize=18)
axs[1, 1].set_ylim([-1, 5])

# Add axis titles for all subplots
for i in range(2):
    for j in range(2):
        axs[i][j].axvline(x=0, color="r", linestyle="-.")
        axs[i][j].set_ylabel("f(x)", fontsize=18)
        axs[i][j].set_xlabel("x", fontsize=18, loc="right")

plt.show()
```
