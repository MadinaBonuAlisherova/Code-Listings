```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, num=200)

# A multimodal function
f = 3 * np.cos(x * x) - np.sin(3 * x) * np.exp(-x)

# Plot formatting
fig = plt.figure(figsize=(15, 15))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(x, f, linewidth=2)
plt.xlabel("x", fontsize=18, loc="right")
plt.ylabel("f(x)", fontsize=18)

# Draw a vertical line at the lower
# and upper bounds of the decision variable

plt.axvline(x=-3, ls="--", color="red")
plt.axvline(x=3, ls="--", color="red")

# plt.savefig("CH01_F03.png", format="png", dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 250)

# The base function
f=(-20*x*x+6200*x-350000)/1000

# Use subplots for multiple graphs
fig, axs = plt.subplots(2, 2, sharex=True, figsize=(20, 20))

# The base function
axs[0, 0].plot(x, f, linewidth=2)
axs[0, 0].set_title("Base function: $(-20*x^2+6200*x-350000)$", fontsize=18)
axs[0, 0].set_ylim([-200, 200])
axs[0, 0].plot(155, 130.50, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green")

# Inverse function
axs[1, 0].plot(x, -f, linewidth=2)
axs[1, 0].set_title("Duality: $-(-20*x^2+6200*x-350000)$", fontsize=18)
axs[1, 0].set_ylim([-200, 200])
axs[1, 0].plot(155, -130.50, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green")

# Stretched function
axs[0, 1].plot(x, f * 0.5, linewidth=2)
axs[0, 1].set_title("Stretched: $(-20*x^2+6200*x-350000)/2$", fontsize=18)
axs[0, 1].set_ylim([-200, 200])
axs[0, 1].plot(155, 65.25, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green")

# Shifted function
axs[1, 1].plot(x, f + 3, linewidth=2)
axs[1, 1].set_title("Shifted: $(-20*x^2+6200*x-350000)+3$", fontsize=18)
axs[1, 1].set_ylim([-200, 200])
axs[1, 1].plot(155, 133.5, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green")

# Add axis titles for all subplots
for i in range(2):
    for j in range(2):
        axs[i][j].axvline(x=155, color="r", linestyle="-.")
        axs[i][j].set_ylabel("f(x)", fontsize=18)
        axs[i][j].set_xlabel("x", fontsize=18, loc="right")

plt.show()

# plt.savefig("CH01_F04.png", format="png", dpi=300)
```
