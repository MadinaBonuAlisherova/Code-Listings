```python
import numpy as np
from scipy.optimize import minimize_scalar


def objective(x):  # A
    return 20 * x * x - 6200 * x + 350000




f = minimize_scalar(objective, method="brent")    #C

opt_x, opt_f = f["x"], -objective(f["x"])

print("Optimal Bike Price: ${:,.2f}".format(opt_x))    #D
print("Resultant Profit f(x): ${:,.2f}".format(opt_f))    #D
```

```python
import matplotlib.pyplot as plt

x = np.arange(0,250,0.1)

# Plot inputs vs. target
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(x, -objective(x), "-")

# Plot the optima
plt.plot([opt_x], [opt_f], "s", color="r")
plt.annotate("(155,130500)",(155,130500))
plt.plot([opt_x], [opt_f], "s", color="r")
plt.annotate("(155,130500)",(155,130500))
plt.plot([opt_x], [opt_f], "s", color="r")
plt.annotate("(155,130500)",(155,130500))

# Add title and axis names
plt.xlabel("Ticket price ($)", fontsize=20)
plt.ylabel("Profit ($)", fontsize=20)
plt.axhline(y=0, ls="-", color="black")
plt.show()
```
