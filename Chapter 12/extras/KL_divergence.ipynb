```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy, norm

# Parameters for the first Gaussian distribution
mu1 = 0  # Mean
sigma1 = 1  # Standard deviation

# Parameters for the second Gaussian distribution
mu2 = 2  # Mean
sigma2 = 1.5  # Standard deviation

# Generate samples from the two Gaussian distributions
x = np.linspace(-10, 10, 100)
p = norm.pdf(x, mu1, sigma1)  # Probability density function for the first distribution
q = norm.pdf(x, mu2, sigma2)  # Probability density function for the second distribution

# Compute the KL divergence
kl_div = entropy(p, q)

# Plotting the two Gaussian distributions and KL divergence
plt.figure(figsize=(8, 6))
plt.plot(x, p, label='p: N(0, 1)', linestyle='dashed',)
plt.plot(x, q, label='q: N(2, 1.5)', linestyle='dotted',)
plt.fill_between(x, 0, np.minimum(p, q), alpha=0.3, color='gray')
plt.title('KL Divergence: {:.4f}'.format(kl_div))
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
# plt.show()
plt.savefig('KL_Divergence.png', dpi=300, bbox_inches='tight', transparent=True)
```
