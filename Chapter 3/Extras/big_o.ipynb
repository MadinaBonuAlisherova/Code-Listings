```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import factorial
import math

x = np.linspace(0.001,20)
a = factorial(x)
b = np.exp2(x)
c = [n**2 for n in x]
d = [n*math.log(n) for n in x]
e = [math.log(n) for n in x]
plt.figure(figsize=(10,6))
plt.plot(x,a,'r',label='$O(n!)$')
plt.plot(x,b,'g',label='$O(2^n)$')
plt.plot(x,c,'b',label='$O(n^2)$')
plt.plot(x,d,'g',linestyle='dashed',label='$O(n log n)$')
plt.plot(x,x,'b',linestyle='dashed',label='$O(n)$')
plt.plot(x,e,'magenta',label='$O(log n)$')
plt.plot(x,[1]*len(x),'r',linestyle='dashed',label='$O(1)$')
plt.xlim(0,20)
plt.ylim(0,30)
plt.xlabel("Input size", labelpad=20)
plt.ylabel("Time", labelpad=20)
plt.xticks(range(0,21,2))
plt.legend()

plt.savefig("Fig_3_14_Big_O.png", format="png", dpi=1200)
plt.savefig("FFig_3_14_Big_O.pdf", format="pdf", dpi=1200)

plt.show()
```
