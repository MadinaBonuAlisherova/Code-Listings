```python
import osmnx as ox
ox.config(requests_kwargs={"verify":False})

G = ox.graph_from_place("Toronto")
```

```python
from tqdm.notebook import tqdm
import math
def Dijkstra(G, origin, destination):
    # Using a set here avoids the problem with self loops
    seen = set()
    shortest_dist = {osmid: math.inf for osmid in G.nodes()}
    unrelaxed_nodes = [Node(graph=G, osmid=osmid) for osmid in G.nodes()]
    shortest_dist[origin.osmid] = 0
    found = False
    pbar = tqdm()
    while len(unrelaxed_nodes) > 0 and not found:
        pbar.set_description(f"Explored: {len(seen)} | Frontier: {len(unrelaxed_nodes)}")
        node = min(unrelaxed_nodes, key=lambda node: shortest_dist[node.osmid])
        # relaxing the node, so this node's value in shortest_dist is the shortest distance between the origin and destination
        unrelaxed_nodes.remove(node)
        seen.add(node.osmid)
        # if the destination node has been relaxed then that is the route we want
        if node == destination:
            found = True
            continue
        # otherwise, let's relax edges of its neighbours
        for child in node.expand():
            # skip self-loops
            if child.osmid in seen:
                continue
            child_obj = next(
                (node for node in unrelaxed_nodes if node.osmid == child.osmid), None
            )
            child_obj.distance = child.distance
            distance = shortest_dist[node.osmid] + child.distance
            if distance < shortest_dist[child_obj.osmid]:
                shortest_dist[child_obj.osmid] = distance
                child_obj.parent = node
    return list(seen)
```

```python
from tqdm.notebook import tqdm
import math
# define destination and origin for the backwards expansion

origin = Node(G,262452973)
destination = Node(G,9084095821)
destination_b = origin
origin_b = destination

frontier_f = set()
frontier_b = set()
frontier_f.add(origin)
frontier_b.add(destination)

explored_f = set()
explored_b = set()

shortest_dist_f = {osmid: math.inf for osmid in G.nodes()}
shortest_dist_b = {osmid: math.inf for osmid in G.nodes()}

shortest_dist_f[origin] = 0
shortest_dist_b[destination] = 0

collide = False
found = False
altr_expand = False # to alternate between front and back
pbar = tqdm()
while frontier_f and frontier_b and not collide and not found:
    pbar.set_description(f"Explored F: {len(explored_f)} | Frontier F: {len(frontier_f)} | Explored B: {len(explored_b)} | Frontier F: {len(frontier_b)} | Collide: {collide} | Altr_Expand: {altr_expand}")
    if altr_expand:
        # remove node_f from frontier_f to expand it
        node = min(frontier_f, key=lambda node: shortest_dist_f[node.osmid])
        frontier_f.remove(node)
        explored_f.add(node)
        
        for child in node.expand():
            if child in explored_f: continue
            
            if child == origin_b:
                found = True
                break

            # checking for collision with the target expansion
            if child in explored_b:
                collide = True
                break

            frontier_f.add(child)
        altr_expand = False
    else:
        # remove node_b from frontier_b to expand it
        node = min(frontier_b, key=lambda node: shortest_dist_b[node.osmid])

        frontier_b.remove(node)
        explored_b.add(node)
        
        for child in node.expand():
            if child in explored_b: continue
            if child == destination_b:
                found = True
                break

            if child in explored_f:
                collide = True
                break

            frontier_b.add(child)
        altr_expand = True
```

```python
def node_c(forward,backward,node):
    if node in forward: return 'blue'
    if node in backward: return 'red'
    return 'black'
```

```python
def edge_c(forward,backward,edge):
    start, end, key = edge
    if start in forward and end in forward: return 'blue'
    if start in backward and end in backward: return 'red'
    return 'black'
```

```python
forward_n = [n.osmid for n in explored_f]
backward_n = [n.osmid for n in explored_b]
```
